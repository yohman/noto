```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Create a new figure and axis
fig, ax = plt.subplots()

# Plot gdf3 in red
gdf3.plot(ax=ax, color='whitesmoke', edgecolor='gray', linewidth=0.2)

# Plot gdf1 in blue
gdf1.plot(ax=ax, color='blue')

# Plot gdf2 in green
gdf2.plot(ax=ax, color='green')

gdfs = [gdf1, gdf2]

# Merge the GeoDataFrames
gdf1_2 = gpd.GeoDataFrame(pd.concat(gdfs, ignore_index=True))

# Optional: Reset the index of the merged GeoDataFrame
gdf1_2.reset_index(drop=True, inplace=True)

# Set the x and y axis limits to the extent of the merged GeoDataFrames
x_min, y_min, x_max, y_max = gdf1_2.total_bounds
x_min -= 0.01
y_min -= 0.01
x_max += 0.01
y_max += 0.01
ax.set_xlim(x_min, x_max)
ax.set_ylim(y_min, y_max)

# Show the plot
plt.show()
```

```python
gdf1 = gdf1.to_crs(4326)
gdf2 = gdf2.to_crs(4326)
gdf3 = gdf3.to_crs(4326)

gdfs = [gdf1, gdf2, gdf3]

merged_gdf = gpd.GeoDataFrame(pd.concat(gdfs, ignore_index=True))

merged_gdf.reset_index(drop=True, inplace=True)
```
